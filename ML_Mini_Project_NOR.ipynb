```python
# ==============================
# NOR Gate using Perceptron
# ==============================

import numpy as np
import matplotlib.pyplot as plt

# ------------------------------
# 1. Dataset (NOR truth table)
# ------------------------------
X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])   # inputs
y = np.array([1,0,0,0]) # outputs for NOR

# ------------------------------
# 2. Initialize parameters
# ------------------------------
np.random.seed(42)
weights = np.random.randn(2)   # two inputs
bias = np.random.randn()
learning_rate = 0.1
epochs = 20

# ------------------------------
# 3. Training using Perceptron rule
# ------------------------------
for epoch in range(epochs):
    for i in range(len(X)):
        linear_output = np.dot(X[i], weights) + bias
        y_pred = 1 if linear_output >= 0 else 0
        error = y[i] - y_pred

        # update rule
        weights += learning_rate * error * X[i]
        bias += learning_rate * error

    print(f"Epoch {epoch+1}: weights={weights}, bias={bias}")

# ------------------------------
# 4. Predictions
# ------------------------------
print("\nFinal Predictions:")
for i in range(len(X)):
    linear_output = np.dot(X[i], weights) + bias
    y_pred = 1 if linear_output >= 0 else 0
    print(f"Input: {X[i]} -> Predicted: {y_pred}, Actual: {y[i]}")

# ------------------------------
# 5. Visualization
# ------------------------------
plt.figure(figsize=(6,6))

# Plot data points
for i in range(len(X)):
    if y[i] == 1:
        plt.scatter(X[i][0], X[i][1], color='green', marker='o', s=120, label="Class 1" if i==0 else "")
    else:
        plt.scatter(X[i][0], X[i][1], color='red', marker='x', s=120, label="Class 0" if i==1 else "")

# Decision boundary
x_vals = np.linspace(-0.5, 1.5, 100)
y_vals = -(weights[0] * x_vals + bias) / weights[1]
plt.plot(x_vals, y_vals, 'b--', label="Decision Boundary")

plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Perceptron Learning NOR Gate")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Perceptron

# Recreate NOR dataset
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([1,0,0,0])   # NOR truth table

# --- Histogram of Y_NOR ---
plt.figure(figsize=(5,4))
plt.hist(y, bins=[-0.5, 0.5, 1.5], rwidth=0.8, color="#5dade2", edgecolor="black")
plt.xticks([0,1])
plt.xlabel("Output (Y_NOR)")
plt.ylabel("Frequency")
plt.title("Fig 1. Histogram of Y_NOR")
plt.show()

# --- Loss Curve ---
errors = []
model = Perceptron(max_iter=1, eta0=0.1, random_state=42, warm_start=True)

# train for 20 epochs manually
for epoch in range(20):
    model.fit(X, y)
    errors.append(np.sum(model.predict(X) != y))

plt.figure(figsize=(6,4))
plt.plot(range(1, len(errors)+1), errors, marker="o", color="#e74c3c")
plt.xlabel("Epoch")
plt.ylabel("Number of misclassifications")
plt.title("Fig 3. Perceptron Training Loss (NOR)")
plt.show()
```
